```python
# PDF reader
def read_pdf(fileloc):
    import PyPDF2
    out=[]
    #pdffile=PyPDF2.PdfFileReader(open('fileloc', mode='rb'))
    pdfFileObject = open(fileloc, 'rb')
    pdfReader = PyPDF2.PdfFileReader(pdfFileObject)
    count = pdfReader.numPages
    for i in range(count):
        page = pdfReader.getPage(i)
        output = page.extractText()
        out.append(output)
    add = ''
    result = add.join(out)
        
    return result
```

```python
# txt reader
def read_txt(file):
    file=open(file,encoding='utf-8')
    file=file.read()
    
    return file
    
```

```python
# Lowercase and Remove Stop Word
def lowercase_stopword(file):
    import spacy
    from nltk.tokenize import word_tokenize

    sp=spacy.load('en_core_web_sm')
    StopWord=sp.Defaults.stop_words
    file_token=word_tokenize(file)

    file_NoStopWord=[word for word in file_token if not word in StopWord]
    file1=[string.lower() for string in file_NoStopWord]
    file1=[string.strip("@#$%^&*(,./?][)]:;''~\n") for string in file1]
    return file1
```

```python
def remove_space(f):
        
    # Remove space
    for i in f:
        if i == '':
            f.remove('')
    return f
```

```python
def Lemmatize(filelist):
    import nltk 
    from nltk.stem import WordNetLemmatizer
    lemmatizer=WordNetLemmatizer()
    lemmatized_output=[lemmatizer.lemmatize(word) for word in filelist]
    return lemmatized_output
```

```python
def feature_extraction(filelist):
    from sklearn.datasets import fetch_20newsgroups
    from sklearn.feature_extraction.text import CountVectorizer
    from sklearn.feature_extraction.text import TfidfVectorizer
    from sklearn.cluster import KMeans
    from sklearn import metrics
    count_vec = CountVectorizer(max_features=30, stop_words='english') #how about using TfidfVectorizer
    X = count_vec.fit_transform(filelist)
    result=X.getrow(0).toarray()
    #vector_input_data = count_vec.transform([filelist]).toarray()[0]
    return count_vec.get_feature_names()
```

```python
def Build_WordCloud(filelist):
    # Import the wordcloud library
    from wordcloud import WordCloud
    # Join the different processed titles together.
    long_string = ','.join(filelist)#list(papers['paper_text_processed'].values))
    # Create a WordCloud object
    wordcloud = WordCloud(background_color="white", max_words=5000, contour_width=3, contour_color='steelblue')
# Generate a word cloud
    wordcloud.generate(long_string)
# Visualize the word cloud
    return wordcloud.to_image()
```

```python
def TopicModel(COP_corpus):

    import nltk
    nltk.download('stopwords')

    import gensim
    from gensim.utils import simple_preprocess
    import nltk
    from nltk.corpus import stopwords
    stop_words = stopwords.words('english')
    stop_words.extend(['from', 'subject', 're', 'edu', 'use'])
    def sent_to_words(sentences):
        for sentence in sentences:
            yield(gensim.utils.simple_preprocess(str(sentence), deacc=True))
    def remove_stopwords(texts):
        return [[word for word in simple_preprocess(str(doc)) 
                 if word not in stop_words] for doc in texts]
    COP_corpus = list(sent_to_words(COP_corpus))
    COP_corpus = remove_stopwords(COP_corpus)

    import gensim.corpora as corpora
    id2word = corpora.Dictionary(COP_corpus)
    texts = COP_corpus
    corpus = [id2word.doc2bow(text) for text in texts]

    from pprint import pprint
    num_topics = 3
    lda_model = gensim.models.LdaMulticore(corpus=corpus,
                                           id2word=id2word,
                                           num_topics=num_topics)
    pprint(lda_model.print_topics())
    doc_lda = lda_model[corpus]

    from pyLDAvis import gensim_models
    import pickle 
    import pyLDAvis
    import os
    pyLDAvis.enable_notebook()
    LDAvis_data_filepath = os.path.join(os.getcwd()+ "/LDATopics"+str(num_topics))
    if 1 == 1:
        LDAvis_prepared = gensim_models.prepare(lda_model, corpus, id2word)
        with open(LDAvis_data_filepath, 'wb') as f:
            pickle.dump(LDAvis_prepared, f)
    with open(LDAvis_data_filepath, 'rb') as f:
        LDAvis_prepared = pickle.load(f)
    pyLDAvis.save_html(LDAvis_prepared, os.getcwd()+ "/COP20"+'.html')
    
    return LDAvis_prepared
```
